```python
stu_list=[]
courses=tuple(["人工智能","马哲","英语","高等数学"])
```

```python
class Student():
    def __init__(self,stu_no,name):
        self.stu_no=stu_no
        self.name=name
    def input_Scores(self):
        score_dict={}
        for key in courses:
            score = int(input(f"请输入{key}的成绩:"))
            score_dict[key]=score
        self.scores=score_dict
        return score_dict
    def getAverage(self):
        sum =0
        for key,value in self.scores.items():
            sum =sum + value
        average =sum/len(self.scores)
        return average
        
```

```python
def AddNewStudent():
    no = input("请输入学号:")
    name = input("请输入姓名:")
    newstu= Student(no,name)
    newstu.input_Scores()
    stu_list.append(newstu)
    print(f"新加入学生的平均成绩为{newstu.getAverage()}")
```

```python
def list_stu():
    for stu in stu_list:
        print(f"学号：{stu.stu_no}姓名{stu.name},成绩如下：\n")
        for key,value in stu.scores.items():
            print(f"{key}：{value}")
```

```python
def computecourseaverage():
    course_average={}
    for course in courses:
        course_average[course]=0
    for stu in stu_list:
        for key,value in stu.scores.items():
            course_average[key] += value
    for key,value in course_average.items():
        course_average[key]=value/len(stu_list)
    return course_average
        
```

```python
while(1):
    choice = int(input('''输入你要进行的选择:0.退出程序
                                       1.创建一个新的学生，
                                       2.计算现有学生的各科平均成绩，
                                       3.列出所有学生各科成绩,
                                       '''))
    if choice == 0:
        break
    elif choice == 1:
        AddNewStudent()
    elif choice == 2:
        course_average=computecourseaverage()
        for key,value in course_average.items():
            print(f"{key}的平均成绩为{value}")
    elif choice ==3:
        list_stu()
    else:
        print("输入有误，请重新输入")
        
```
